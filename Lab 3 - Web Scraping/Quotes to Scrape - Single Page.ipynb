```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
page = requests.get('https://quotes.toscrape.com')
page
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
soup
```

```python
quote_elements = soup.find_all('div', class_='quote')
quote_elements
```

```python
# List to store quotes
quotes = []
```

```python
for quote_element in quote_elements:
        text = quote_element.find('span', class_='text').text
        author = quote_element.find('small', class_='author').text
        tags = ', '.join(tag.text for tag in quote_element.find_all('a', class_='tag'))
        quotes.append({'Text': text, 'Author': author, 'Tags': tags})
```

```python
quotes
```

```python
# Remove quotation marks from Text field
for quote in quotes:
    quote['Text'] = quote['Text'].strip('“”')

# Save quotes to CSV file
with open('quotes1.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=['Text', 'Author', 'Tags'])
    writer.writeheader()
    writer.writerows(quotes)
```
